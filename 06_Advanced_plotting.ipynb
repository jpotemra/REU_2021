# Advanced plotting

By now you should be somewhat comfortable making plots with python/matplotlib.  Let's now look at more plotting options and reading in actual data, combining all that we've learned so far.

In this example, we will use pandas to read an ASCII data set, in fact a comma-separated-value data set that is available via the web.  Let's first make the plot then have a look at its awesomeness.

```python
# as usual, import needed packages and set in-line display
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# read data from (remote) file
midwest = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")

# Plot as scatter plot with circles sized by a value
fig = plt.figure(figsize=(14, 7), dpi= 80, facecolor='w', edgecolor='k')    
plt.scatter('area', 'poptotal', data=midwest, s='dot_size', c='popdensity', cmap='Reds', edgecolors='black', linewidths=.5)
plt.title("Bubble Plot of PopTotal vs Area\n(color: 'popdensity' & size: 'dot_size' - both are numeric columns in midwest)", fontsize=16)
plt.xlabel('Area', fontsize=18)
plt.ylabel('Poptotal', fontsize=18)
plt.colorbar()
```

Okay, the first thing to note is we read the data from a web-accessible data set.  Note we are not actually transferring the data set from there to her.  Instead, we assign the contents to the variable.  How can we tell what type of variable we now have?  And what's in it?

So, we've created a variable called "midwest" and it happens to be a pandas DataFrame.  This of this like a table.  In fact when looking at it, we see that we have column headings, mixed text and numbers, etc.  All the things that caused problems in other tools.  Python/pandas read this in no problem.

Next we make a scatter plot.  In detail, we have:

plt.scatter('area', 'poptotal', data=midwest, s='dot_size', c='popdensity', cmap='Reds', edgecolors='black', linewidths=.5)

Note here instead of specifying an x-axis variable and a y-axis variable, we simply need to provide (in quotes) the column heading that we want to plot (!).  In this example, the x-axis is the column titled "area" and the y-axis is the column titled "poptotal".

Next, we make the color of each data point linked to a third variable, namely "popdensity", and the size based on a 4th variable "dot_size" (note, this final variable is actualy a variable name in the data set, it doesn't seem obvious what this represents).

Let's now try gussy this up a bit.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

# Import data
midwest = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/midwest_filter.csv")

# Plot
fig = plt.figure(figsize=(14, 9), dpi= 80, facecolor='w', edgecolor='k')    
colors = plt.cm.tab20.colors
categories = np.unique(midwest['category'])
for i, category in enumerate(categories):
    plt.scatter('area', 'poptotal', data=midwest.loc[midwest.category==category, :], s='dot_size', c=colors[i], label=str(category), edgecolors='black', linewidths=.5)

# Legend for size of points
for dot_size in [100, 300, 1000]:
    plt.scatter([], [], c = 'k', alpha=0.5, s=dot_size, label=str(dot_size) + ' TotalPop')
plt.legend(loc='upper right', scatterpoints=1, frameon=False, labelspacing=2, title='Saravana Stores', fontsize=8)
plt.title("Bubble Plot of PopTotal vs Area\n(color: 'category' - a categorical column in midwest)", fontsize=18)
plt.xlabel('Area', fontsize=16)
plt.ylabel('Poptotal', fontsize=16)
```
